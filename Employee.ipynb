```python
import pandas as pd
import matplotlib.pyplot as plt

# Upload file
from google.colab import files
uploaded = files.upload()
```

```python
# Ganti nama file sesuai file yang kamu upload
file_name = 'employee_data.xls'  # Ganti sesuai nama file kamu

# Baca data dari Excel (ambil sheet pertama)
df = pd.read_excel(file_name)

# Tampilkan nama kolom
print("Nama kolom dalam data:")
print(df.columns)
```

```python
# Hitung jumlah tiap jenis status (izin/cuti/resign, dll)
izin_cuti = df['TerminationDescription'].value_counts()

# Filter hanya yang mengandung kata 'Leave', 'Izin', atau 'Cuti'
izin_cuti_filtered = izin_cuti[izin_cuti.index.str.contains("leave|izin|cuti", case=False)]

# Tampilkan datanya
print(izin_cuti_filtered)
```

```python
plt.figure(figsize=(8, 5))
izin_cuti_filtered.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Perbandingan Jumlah Izin vs Cuti')
plt.xlabel('Jenis')
plt.ylabel('Jumlah Karyawan')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
